```python
X = df1.drop('Entity',axis=1)
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
from math import sqrt
sqrt(11912/2)
```

```python
km = KMeans(n_clusters=3)
km.fit(X)
pred = km.predict(X)
pred
```

```python
df1['Cluster'] = pred
```

```python
df1['Cluster'].value_counts()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.scatterplot(data=df1,x='Rice ',y='Maize',palette='winter')
plt.subplot(1,2,2)
sns.scatterplot(data=df1,x='Rice ',y='Maize',palette='winter',hue='Cluster',size='Cluster') 
plt.tight_layout()
plt.show()
```
